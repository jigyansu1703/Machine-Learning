```python
#Merge/join >> merge based on common column, inner merge (intersection)
# >> merged based on the left dataframe, left merge (All the element of data a will be in the merge
#>> right merge>> all the element of right data frame (B) will be in the merge
#All the elements of both the data frames>> outer the data frame merge
#cross merge >> every permutation of rows are there
```

```python
#Merge
df1 = pd.DataFrame({'Key1':[1,2,3,4,5],
                    'Key2':[4,5,6,7,8],
                    'key3':[3,46,5,6,6]
})
df1
```

```python
#Merge
df2 = pd.DataFrame({'Key1':[1,2,3,4,5],
                    'Key4':[4,56,6,77,8],
                    'key5':[3,46,5,6,66]
})
df2
```

```python
#Inner merge
pd.merge(df1,df2, how = 'inner')
```

```python
pd.merge(df1,df2, how = 'left')
```

```python
pd.merge(df1,df2, how = 'right')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, how = 'outer') #all the comman element of the column from the both the dataframe
```

```python
pd.merge(df1,df2, how = 'cross')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = "left", left_on = "Key2", right_on = "Key4")
```
