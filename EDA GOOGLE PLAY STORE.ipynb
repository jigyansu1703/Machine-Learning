```python
#1MB = 1024 kb
```

```python
def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M",""))
        res = res * 1024
        return res
    elif str(item)[-1] == "M":
        res = float(str(item).replace("M",""))
        res = res * 1024
        return res
    else:
        return str(np.nan)
```
