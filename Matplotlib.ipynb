```python
#pie chart use to under stand the percentage distribution of data
langs = ["c","python","java"]
students = [20,100,40]

plt.pie(students,labels = langs, autopct = "%1.1f%%")
plt.title("P")
plt.show()
```

```python
#focus on one segment
explode = (0.0,0.1,0.1)
langs = ["c","python","java"]
students = [20,100,40]
plt.pie(students,labels = langs, autopct = "%1.1f%%",explode = explode)
plt.title("Pie chart")
plt.show()
```

```python
explode = (0.0,0.05,0.05)
langs = ["c","python","java"]
students = [20,100,40]
plt.pie(students,labels = langs, autopct = "%1.1f%%",explode = explode, shadow = True)
plt.title("Pie chart")
plt.show()
```

```python
#Area plot >> variables changed over time
#An area chart is a visual representation of data that utilizes both lines and filled areas to convey information. 
#This type of chart is particularly effective in showcasing data trends and variations over a specified period 
#or across different categories.
```

```python
years = [2021,2018,2022,2023,2017]
category1 = [20,25,30,28,35]
category2 = [10,15,20,18,25]
category3 = [5,8,10,9,15]
```

```python
plt.stackplot(years, category1,category2,category3, labels = ["category1","category2","category3"],colors = ['skyblue','lightgreen','orange'])
plt.xlabel("years")
plt.ylabel("Values")
plt.title("Area plot")
plt.legend()
plt.show()
```
