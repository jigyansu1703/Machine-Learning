```python
#to check the run time of a code
start = pd.Timestamp.now()
for i in range(1000):
    pass
end = pd.Timestamp.now()
print("Time taken for the execution of code", end-start)
```

```python
#another way
import time
start = time.time()
start
```

```python
#another way
import time
start = time.time()
for i in range(1000):
    pass
end = time.time()
print("Time taken for the execution of code", end-start)
```

```python
#one more way to check executtion time
import time
start = time.perf_counter()
for i in range(1000):
    pass
end = time.perf_counter()
print("Time taken for the execution of code", end-start)
```
